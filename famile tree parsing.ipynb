```python
# Define the family tree as a nested dictionary
family_tree = {
    "Alice": {"parents": ["John", "Mary"], "children": ["Bob", "Carol"]},
    "John": {"parents": [], "children": ["Alice"]},
    "Mary": {"parents": [], "children": ["Alice"]},
    "Bob": {"parents": ["Alice", "David"], "children": []},
    "Carol": {"parents": ["Alice", "David"], "children": []},
    "David": {"parents": [], "children": ["Bob", "Carol"]}
}

# Function to get parents of a person
def get_parents(person, family_tree):
    return family_tree.get(person, {}).get("parents", [])

# Function to get children of a person
def get_children(person, family_tree):
    return family_tree.get(person, {}).get("children", [])

# Function to get siblings of a person
def get_siblings(person, family_tree):
    parents = get_parents(person, family_tree)
    siblings = set()
    for parent in parents:
        for child in get_children(parent, family_tree):
            if child != person:
                siblings.add(child)
    return list(siblings)

# Function to get grandparents of a person
def get_grandparents(person, family_tree):
    grandparents = set()
    parents = get_parents(person, family_tree)
    for parent in parents:
        grandparents.update(get_parents(parent, family_tree))
    return list(grandparents)

# Example usage
person = "Bob"
print(f"Parents of {person}: {get_parents(person, family_tree)}")
print(f"Children of {person}: {get_children(person, family_tree)}")
print(f"Siblings of {person}: {get_siblings(person, family_tree)}")
print(f"Grandparents of {person}: {get_grandparents(person, family_tree)}")
```
